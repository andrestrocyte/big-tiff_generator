```python
"""
Created on Fri Aug 18 20:26:41 2023

@author: andrestrocyte based on mahnmath code
"""


import os
import re
from tkinter import Tk
from tkinter.filedialog import askdirectory
import tifffile

Tk().withdraw()
path = askdirectory(title="Select a folder to process TIFF files")

# Define the output TIFF file
output_tiff_path = os.path.join(path, 'bigtiff_combined.tif')

# Flag to indicate if the output TIFF file has been created
output_tiff_created = False

for root, _, files in os.walk(path):
    for file in files:
        if re.match(r'.*_Cycle00001_Ch[12]_000001.ome.tif', file):
            file_path = os.path.join(root, file)

            print(f'Series {file_path} being processed')

            if not output_tiff_created:
                # Create the output TIFF file with the metadata from the first file
                with tifffile.TiffWriter(output_tiff_path, bigtiff=True) as tif:
                    img = tifffile.imread(file_path)
                    tif.save(img)

                output_tiff_created = True
            else:
                # Append frames from the current file to the existing output TIFF file
                with tifffile.TiffWriter(output_tiff_path, bigtiff=True, append=True) as tif:
                    img = tifffile.imread(file_path)
                    tif.save(img)

print(f'Combined TIFF file created at {output_tiff_path}')
```
